```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1) 
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
w = np.array([1,1])

z = np.dot(X, w) - 3
print(z)
y_pred = sigmoid(z)
print(y_pred)

decision_boundary = 3 - w[1] * X[:,1]

plt.scatter(X[y.ravel() == 0][:, 0], X[y.ravel() == 0][:, 1], marker='o', c='blue', label="y=0")
plt.scatter(X[y.ravel() == 1][:, 0], X[y.ravel() == 1][:, 1], marker='x', c='red', label="y=1")

plt.plot(X[:, 1], decision_boundary, 'g-', label="decision boundary")


plt.axis([0, 4, 0, 3.5])
plt.ylabel('$x_1$')
plt.xlabel('$x_0$')
plt.legend()
plt.show()
```
