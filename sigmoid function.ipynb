The formula for the sigmoid function is:

$$
g(z) = \frac{1}{1 + e^{-z}}
$$

In the case of a single example, $z$ is scalar.<br>
in the case of multiple examples, $z$ may be a vector consisting of $m$ values, one for each example.

```python
def sigmoid(z):
    g = 1 / (1 + np.exp(-z))

    return g
```

```python
z_tmp = np.arange(-10,11)

y = sigmoid(z_tmp)

fig,ax = plt.subplots(1,1,figsize=(5,3))
ax.plot(z_tmp, y, c="b")

ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
```

```python
# Apply logistic regression to the categorical data example of tumor classification.
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])

w_in = np.zeros((1))
b_in = 0
```

```python
y_hat = sigmoid(w_in*x_train + b_in)

plt.scatter(x_train[y_train == 0], y_train[y_train == 0], marker='o', c='blue', label="y=0 (circle)")
plt.scatter(x_train[y_train == 1], y_train[y_train == 1], marker='x', c='red', label="y=1 (cross)")
plt.plot(x_train, y_hat)

plt.plot(x_train, y_hat, c='g', label='Sigmoid function')

plt.xlabel("Tumor size")
plt.ylabel("y")
plt.legend()
plt.show()
```
